```python
trades = conn.raw_sql("""SELECT *
                            FROM taqm_2014.ctm_20140102
                            WHERE sym_root = 'BCS'
                            """, date_cols=['datadate'])
```

```python
print(trades.shape)
print(trades.columns)
trades.head()
```

```python
import pandas as pd

# Create list that will become LOBk
lobk_list = []
bids, asks = {}, {}

# loop through all elements in the quotes
for index, row in quotes.iterrows():
    if index % 100000 == 0:
        print(index)
    # For each quote, access the price and volume for both bid and ask if applicable,
    # and add the volumes to two dictionaries for bids and asks with prices as keys and volume as the value
    if row['bidsiz'] is not None and row['bidsiz'] != 0:
        bids[row['bid']] = bids.get(row['bid'], 0) + row['bidsiz']
    
    if row['asksiz'] is not None and row['asksiz'] != 0:
        asks[row['ask']] = asks.get(row['ask'], 0) + row['asksiz']

    # Once done, compare timestamp of current quote with previous and reduce dictionary values
    # by the quantity of trades that occurred between each quote update
    if index != 0:
        rel_trades = trades[trades['time_m'] <= quotes.iloc[index]['time_m']]
        trades = trades.iloc(rel_trades.shape[1])

        if not rel_trades.empty:
            for _, trd_row in rel_trades.iterrows():
                if trd_row['price'] in bids:
                    bids[trd_row['price']] = max(0, bids[trd_row['price']] - trd_row['size'])
                if trd_row['price'] in asks:
                    asks[trd_row['price']] = max(0, asks[trd_row['price']] - trd_row['size'])

    # Sort and slice bids and asks to get top 10
    top_10_bids = dict(sorted(bids.items(), key=lambda x: -x[0])[:10])  # Descending order
    top_10_asks = dict(sorted(asks.items(), key=lambda x: x[0])[:10])   # Ascending order

    # Append the current state of the top 10 levels for the limit order book (done quickly, need to reformat)
    lobk_entry = {
        'time_m': row['time_m'],
        'bids': top_10_bids,
        'asks': top_10_asks
    }
    lobk_list.append(lobk_entry)

# Convert list to DataFrame and save to CSV
lobk_df = pd.DataFrame(lobk_list)
lobk_df.to_csv("LOBk_data.csv", index=False)
```
