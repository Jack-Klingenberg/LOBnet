```python
quotes = conn.raw_sql(f"""SELECT * FROM taqm_2014.cqm_20140102 WHERE sym_root = 'BCS'; """)
trades = conn.raw_sql(f"""SELECT * FROM taqm_2014.ctm_20140102 WHERE sym_root = 'BCS'; """)
```

```python
trades["size"].max()
```

```python
# Create list that will become LOBk

# loop through all elements in the quotes
#   for each quote acess the price and volume for both bid and ask if applicable and add the volumes to two dictionaries
#   for bids and asks with prices as keys and volume as the value

#   Once done, compare timestamp of current quote with previous and reduce dictionary values by quantity of trades that occured between each quote update

#   Once trades are accounted for insert the N levels into the OBk list

# Exit loop and format dataframe for LOBk
# Write dataframe to a CSV for each day

# price to volume
bids = {}
asks = {}
lobk = []
trades = trades.sort_values(by='time_m')
# initialize the sliding window
prev_trade_timestamp = None
next_trade_timestamp = None

for index, row in quotes.iterrows():
    # access the price and volume for both bid and ask
    ask, ask_size = row["ask"], row["asksiz"]
    bid, bid_size = row["bid"], row["bidsiz"]

    # add the volumes to two dictionaries
    bids[bid] = bid_size + bids.get(bid, 0)
    asks[ask] = ask_size + asks.get(ask, 0)

    curr_timestamp = row["time_m"]
    # initialize the window on the first iteration
    if prev_trade_timestamp is None and next_trade_timestamp is None:
        next_trade_candidates = trades[trades["time_m"] >= curr_timestamp]
        # print(curr_timestamp)
        # print(next_trade_candidates)
        if not next_trade_candidates.empty:
            next_trade_timestamp = next_trade_candidates.iloc[0]["time_m"]
        else:
            break

    # filter trades between prev_trade_timestamp and next_trade_timestamp
    trades_between = trades[(trades["time_m"] >= prev_trade_timestamp) & (trades["time_m"] <= next_trade_timestamp)]

    # reduce dictionary values by quantity of trades that occured between each quote update
    for trade_index, trade_row in trades_between.iterrows():
        trade_price = trade_row["price"]
        trade_size = trade_row["size"]
        if trade_price in bids:
            bids[trade_price] -= trade_size
        if trade_price in asks:
            asks[trade_price] -= trade_size
    
    # update sliding window
    prev_trade_timestamp = next_trade_timestamp
    next_trade_candidates = trades[trades["time_m"] >= next_trade_timestamp]
    if not next_trade_candidates.empty:
        next_trade_timestamp = next_trade_candidates.iloc[0]["time_m"]
    else:
        break
    
    # add to lobk
    lobk.append({
        "time_m": curr_timestamp,
        "bids": bids,
        "asks": asks
    })
    # Remove processed trades from the DataFrame to make future iterations faster
    trades = trades[trades["time_m"] > prev_trade_timestamp]
# trades
```
